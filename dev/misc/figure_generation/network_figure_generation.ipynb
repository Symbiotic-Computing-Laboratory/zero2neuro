```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
import keras

#################################################################
# Default plotting parameters
FIGURESIZE=(10,6)
FONTSIZE=18

plt.rcParams['figure.figsize'] = FIGURESIZE
plt.rcParams['font.size'] = FONTSIZE

plt.rcParams['xtick.labelsize'] = FONTSIZE-2
plt.rcParams['ytick.labelsize'] = FONTSIZE-2

#################################################################
```

```python
def generate_figure(x,
                    y,
                    xlabel='net input',
                    ylabel='activation',
                    fname=None,
                    title=None,
                    linewidth=2):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    if title is not None:
        plt.title(title)

    if fname is not None:
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        
```

```python
# Sigmoid

x = np.linspace(-6, 6, 300)
y1 = keras.activations.sigmoid(x).numpy()
generate_figure(x,y1, fname='sigmoid.png', title='Sigmoid')
```

```python
# Tanh

x = np.linspace(-6, 6, 300)
y2 = keras.activations.tanh(x).numpy()
generate_figure(x,y2, fname='tanh.png', title='Tanh')
```

```python
# RELU

x = np.linspace(-6, 6, 300)
y3 = keras.activations.relu(x).numpy()
generate_figure(x,y3, fname='relu.png', title='Relu')
```

```python
# elu

x = np.linspace(-6, 6, 300)
y4 = keras.activations.elu(x).numpy()
generate_figure(x,y4, fname='elu.png', title='Elu')
```

```python
# elup1

x = np.linspace(-6, 6, 300)
y5 = keras.activations.elu(x).numpy()+1
generate_figure(x,y5, fname='elup1.png', title='Elup1')
```

```python
# gelu

x = np.linspace(-6, 6, 300)
y6 = keras.activations.gelu(x).numpy()
generate_figure(x,y6, fname='gelu.png', title='Gelu')
```
